### Boxplot
Visualizaci√≥n estad√≠stica que resume la distribuci√≥n de una variable num√©rica, mostrando su mediana, cuartiles, rango y posibles valores at√≠picos (outliers).

Es una herramienta esencial para comparar distribuciones entre grupos.

![Boxplot](./assets/boxplot.png)

üéØ ¬øPara qu√© sirve?
Comparar distribuciones entre categor√≠as
Ej.: salarios por departamento, tiempos de respuesta por servidor, notas por curso.
Identificar asimetr√≠as
Si la mediana no est√° centrada en la caja, la distribuci√≥n est√° sesgada.
Detectar valores at√≠picos
Puntos aislados que podr√≠an ser errores o casos especiales.
Evaluar dispersi√≥n y rango
Cajas anchas = alta variabilidad; cajas estrechas = datos concentrados.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/SOCR-HeightWeight.csv', usecols=['Height(Inches)'])
df["Height(Inches)"] = df["Height(Inches)"] * 2.54 # Convert height to cm

plt.figure(figsize=(6, 5))
sns.boxplot(y=df["Height(Inches)"])  # ‚Üê reemplaza 'altura' por el nombre real de tu columna
plt.title("Distribuci√≥n de alturas", fontsize=14)
plt.ylabel("Altura (cm)")
plt.show()
```

- Eje Y: Altura en cent√≠metros (de 155 cm a 190+ cm).
- Caja azul: Representa el rango intercuart√≠lico (IQR) ‚Üí el 50% central de los datos.
- L√≠nea horizontal dentro de la caja: Mediana ‚âà 172 cm (el valor que divide los datos en dos mitades iguales).

Bigotes (l√≠neas verticales):
- Extremo superior: m√°ximo no at√≠pico ‚âà 185 cm.
- Extremo inferior: m√≠nimo no at√≠pico ‚âà 160 cm.

Puntos individuales (c√≠rculos):
- En la parte superior: varios valores por encima de 185 cm ‚Üí son outliers superiores.
- En la parte inferior: varios valores por debajo de 160 cm ‚Üí son outliers inferiores.

Centro de los datos:
- La mediana es ~172 cm, lo que sugiere que la mitad de las personas miden menos de 172 cm y la otra mitad m√°s.

Dispersi√≥n:
- El IQR (caja) va aproximadamente desde ~170 cm a ~175 cm, lo que significa que el 50% central de las alturas est√° entre esos valores.
- El rango total (sin outliers) va de ~160 cm a ~185 cm, mostrando una variabilidad moderada.

Valores extremos (outliers):
- Hay m√∫ltiples outliers tanto en la parte alta como en la baja:
- Algunas personas miden m√°s de 185 cm (hasta ~190 cm) ‚Üí muy altas.
- Otras miden menos de 160 cm (hasta ~155 cm) ‚Üí muy bajas.

Esto puede indicar que la muestra incluye personas de diferentes grupos (por ejemplo, hombres y mujeres, o adultos y adolescentes).

Forma impl√≠cita:
- La mediana est√° casi en el centro de la caja ‚Üí sugiere simetr√≠a moderada.
- Pero los outliers en ambos extremos indican que hay una cola larga hacia arriba y hacia abajo.
