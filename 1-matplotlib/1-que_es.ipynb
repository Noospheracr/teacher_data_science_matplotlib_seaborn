##### Importación

```python
import matplotlib.pyplot as plt
```

##### Comandos mínimos

```python
plt.figure(figsize=(14, 4)) #! Dimensiones del gráfico, el primero es el ancho y el segundo es el alto en pulgadas

#! Dibuja una línea que conecta puntos en un plano bidimensional.
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])

plt.title('Título del gráfico')

plt.xlabel('Texto eje X')

plt.ylabel('Texto eje Y')

#! Muestra el gráfico
plt.show()
```
