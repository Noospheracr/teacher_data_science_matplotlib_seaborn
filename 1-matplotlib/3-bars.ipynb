### Gráfico de barras (bar chart)
Cada barra representa una categoría y su altura o longitud muestra el valor numérico asociado a la categoría.

- Comparar métricas por categoría
- Mostrar distribuciones categóricas
- Visualizar rankings
- Comparar valores en el tiempo (cuando el tiempo es categórico)

##### Ranking de emisiones de CO2
[Dataset (fuente: Our World in Data)](https://ourworldindata.org/co2-emissions)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataframe
df = pd.read_csv('data/co2-emissions.csv')
df = df[df["Year"] == 2023]
df.dropna(inplace=True)
df_top = df[['Entity', 'Annual CO₂ emissions']].sort_values('Annual CO₂ emissions', ascending=False).head(10)

# Graficar
plt.figure(figsize=(14, 6))

bars = plt.barh(df_top['Entity'][::-1], df_top['Annual CO₂ emissions'][::-1], color='#e74c3c')  # rojo intenso

# Añadir valores en las barras
for i, v in enumerate(df_top['Annual CO₂ emissions'][::-1]):
    plt.text(v + 100, i, f"{v:,.0f} Mt", va='center', fontsize=11)

plt.title("Top 10 países con mayores emisiones de CO₂ (2023)", fontsize=14, weight='bold')
plt.xlabel("Emisiones de CO₂ (millones de toneladas)")
plt.show()

```
